```python
df['filtered']=df["tweet"].apply(lambda text: " ".join(token.lemma_ for token in nlp(text) if not token.is_punct and not token.is_stop and not token.is_currency
              and not token.is_digit
              and not token.is_oov
              and not token.is_space
              and not token.like_num
              and not token.pos_ == "PROPN"))
```

```python
df['filtered']=df['filtered'].apply(lambda x : " ".join(y.lower() for y in x.split()))
```

```python
import re
df['filtered']=df['filtered'].apply(lambda x: re.sub('[^A-Za-z ]',' ',x))
```

```python
for i in range(len(df)):
    df['filtered'][i]=df['filtered'][i].replace('m','')
```

```python
df['filtered']=df['filtered'].apply(lambda x : [token.text for token in nlp(x)])
```
